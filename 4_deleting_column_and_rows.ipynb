```python
import pandas
fhand = pandas.read_json('5.3 supermarkets.json')
fhand = fhand.set_index('Address')
fhand
```

```python
#delete a column

fhand.drop('City', 1)  # pass 1 as the 2nd argument if we delete a column
```

```python
#delete a row

fhand.drop('332 Hill St', 0)  # pass 0 as the 2nd argument if we delete a row
```

```python
# deleting column by position

fhand.drop(fhand.columns[2:5], 1)
```

```python
# same goes with column

fhand.drop(fhand.index[1:3], 0)
```
