## Adding column to a data frame.
Let's do. here

````python
import pandas
df = pandas.read_csv('5.2 supermarkets.csv')
df = df.set_index('Address')
df
````

**Now we will add a column name 'continent' and add value 'North America' for all of the rows**

````python
df['Continent'] = df.shape[0]*['North America']
````

### Let's intercept the command line.

* ```df["continent"]  ``` will add a column

* Shape of df is (6, 6) ```df.shape[row, column]```

* ```shape[0]``` represents the row

* ```shape[0]*['North America']``` add data to these 6 rows

***Let's print the df now***


````python
df['Continent'] = df.shape[0]*['North America']
````

````python
df
````

### We can modify column using this command

```
df['Continent'] = df['City'] +', ' + 'North America'
```
* we have combined the *City* column and a string *North America* to the **Continent Column**

***Let's print the data frame now***


````python
df['Continent'] = df['City'] +', ' + 'North America'
````

````python
df
````

## Adding a row to the data frame

Let's do it. We've to play a tricks to do that. We'll transpose the columns. Then we'll add a column and transpose it again. for this we'll use ```df.T``` command.

````python
df_r = df.T
````

````python
df_r
````

* **We've transposed the columns to row. No we've to add a new column which we wanted to see in a new row**

````python
df_r['My Address'] = [7, 'My City', 'My State', 'My Country', 'My Name', 'My Employees', 'My Continent']
````

````python
df_r
````

* **Now if we transpose the dataframe again we'll have our new row**

````python
df_r = df_r.T
````

````python
df_r
````

### How to modify a row?

* We have to do it in the same way. We have to modify an existing column after transposing instead of adding a new column

* ie: ```df_r['332 Hill St'] = [7, 'My City', 'My State', 'My Country', 'My Name', 'My Employees', 'My Continent']```

* ```'332 Hill St'``` is a column header after transposing. If we modify it's row value we'll get our modified row


**Let's Do it**

````python
df_r = df_r.T #transposing
````

````python
df_r
````

````python
df_r['332 Hill St'] = [7, 'My City', 'My State', 'My Country', 'My Name', 'My Employees', 'My Continent'] 
#modifying existing column here
````

````python
df_r
````

````python
# Now it's modified. we've to transpose again to get them as row
df_r = df_r.T
````

````python
df_r
````
